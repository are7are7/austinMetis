# Its time to use a gradient calculator

Time to add in a powerful gradient calculator that works well with the numpy library.  [Click here to view the github repo](https://github.com/HIPS/autograd) - download its contents - and toy around.  In particular figure out how to use it to compute derivatives of elementary functions like

- $y=\text{sin}(x)$
- $y=x^p$ where p is a fixed power
- $y=\text{max}(0,x)$

Plot both the original function as well as its derivative to make sure you're using the code correctly.

```python
import matplotlib.pyplot as plt
import autograd.numpy as np  # Thinly-wrapped numpy
from autograd import grad    # The only autograd function you may ever need
```

```python
def func(x):                 # Define a function
    y =np.maximum(x,0)
    return y

# produce gradient
my_grad = grad(func)

# now linspace it and plot
s = np.linspace(-5,5)
g = [func(v) for v in s]
g_p = [my_grad(v) for v in s]
plt.plot(s,g,linewidth=2)
plt.plot(s,g_p,linewidth=2)
plt.legend(['func','derivative'],loc='center left', bbox_to_anchor=(1, 0.5))
# plt.axis('off')
plt.show()
```

```python
def func2(x):
    y = np.sin(x) + np.cos(x)
    return y
def func(x):                 # Define a function
    y = func2(x)**2
    return y

# produce gradient
my_grad = grad(func)

# now linspace it and plot
s = np.linspace(-5,5,500)
g = [func(v) for v in s]
g_p = [my_grad(v) for v in s]
plt.plot(s,g,linewidth=2)
plt.plot(s,g_p,linewidth=2)
# plt.axis('off')
plt.legend(['func','derivative'],loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
def func3(x):
    y = x**3
    return y
def func2(x):
    y = np.sin(func3(x))
    return y
def func(x):                 # Define a function
    y =np.sin(x + func2(x))
    return y

# produce gradient
my_grad = grad(func)

# now linspace it and plot
s = np.linspace(-5,5,500)
g = [func(v) for v in s]
g_p = [my_grad(v) for v in s]
plt.plot(s,g,linewidth=2)
plt.plot(s,g_p,linewidth=2)
plt.axis('off')
plt.legend(['func','derivative'],loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
