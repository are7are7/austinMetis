# Cross-validation: layer 1

In the regression and classification notebooks we saw how each type of supervised learning algorithm - kernels, tree-based learners, and neural networks - have parameter(s) that need to be properly tuned in order for them to provide useful predictive power.  Failure to do th is results in a learner that either **underfits** the data - providing too stiff of a model to explain the underlying data - or **overfits** the data - providing far too complicated a model for the data.

With the simple toy datasets we have been playing with so far this parameter selection can be done 'by hand'.  That is, we can adjust the parameter to its optimal value by visually checking that it does a good job of approximating the underlying data generating function.  However in practice doing this work by hand is impossible.  This is both because typical datasets have too many dimensions to be visualized (so say goodbye to the ability to visually check the efficacy of a paramter choice) and because many kernels, trees, and neural nets have too many parameters to adjust (e.g., modern deep neural nets can easily have thousands or even millions of parameters that need adjusting).  

So we need a way to do this parameter adjustment automatically, regardless of the dimension of the dataset and number of parameters to adjust.  The most common way of doing this is a procedure known as **cross-validation**.  In this notebook you will learn how to use scikit-learn's built in functionality for this procedure for each of the core algorithm types.  The point of this exercise is to get you familiar with the functionality and (scikit) syntax of these fundamental procuedures.  In Layer 2 of the course we will go deeper into exactly how cross-validation works, and write some code to perform cross-validation that mirrors very closely to what you will be using from scikit-learn (or any other cross-validation functionality for that matter).

To help you build your intuition about how cross-validation performs in practice we will stick with our toy datasets for this exercise, so that you can see how well the results work.  In a future notebook in Layer 1 you will then apply cross-validation functionality on a real dataset.

In this short section we will apply cross-validation with the algorithms above to a real credit card approval dataset.  This is a dataset consisting of around 700 datapoints, consisting of a number of features about a potential credit card holder and whether or not this person was approved for a card.  You can find a full description of this data via the link below

https://archive.ics.uci.edu/ml/datasets/Statlog+(Australian+Credit+Approval)

With 10+ input features this data is too high dimensional for us to visualize properly, but we can still tell how well the algorithm performs by examining a classifier's accuracy.  Later today we will learn about more precise measures of how classifier goodness, but for now accuracy will suffice.

```python
# here labels are in the first column, with features the remaining ones
loan_data = np.asarray(pd.read_csv('credit_approval_dataset.csv',delimiter=',',header = None))
labels = loan_data[:,0]         # get the labels
labels.shape = (len(labels),1)
data = loan_data[:,1:]        # get the features
```

### Trying out trees

```python
# import the gradient boosting classifier from scikit
from sklearn.ensemble import GradientBoostingClassifier

# create an instance 
clf = GradientBoostingClassifier()
param_name = 'max_depth'
param_range = np.arange(1,20)

# run our custom cross-validation function for this choice of classifier and param_name
train_errors, test_errors  = simple_cross_validation(data,labels,clf,param_name,param_range)

# plot training and testing errors
plot_cv_scores(train_errors,test_errors,param_range)
```

```python
# pick the best parameter - i.e., the one that minimizes test error
best_param = param_range[np.argmin(test_errors)]
print 'our best cross-validated choice of parameter was: ' + param_name + ' = ' + str(best_param)

# create an instance of our classifier with the cross-validated best choice of parameter
clf = GradientBoostingClassifier(n_estimators = 1,max_depth = best_param)

# fit classifier
clf.fit(data,labels.ravel())

# print out the number of misclassified points
predicted_labels = clf.predict(data)
acc = len(labels) - accuracy_score(labels.ravel(), predicted_labels.ravel(), normalize=False)
print 'Our classifier mislabeled ' + str(acc) + ' of ' + str(len(labels)) + ' points'
```
