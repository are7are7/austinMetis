# Cross-validation: layer 1

In the regression and classification notebooks we saw how each type of supervised learning algorithm - kernels, tree-based learners, and neural networks - have parameter(s) that need to be properly tuned in order for them to provide useful predictive power.  Failure to do th is results in a learner that either **underfits** the data - providing too stiff of a model to explain the underlying data - or **overfits** the data - providing far too complicated a model for the data.

With the simple toy datasets we have been playing with so far this parameter selection can be done 'by hand'.  That is, we can adjust the parameter to its optimal value by visually checking that it does a good job of approximating the underlying data generating function.  However in practice doing this work by hand is impossible.  This is both because typical datasets have too many dimensions to be visualized (so say goodbye to the ability to visually check the efficacy of a paramter choice) and because many kernels, trees, and neural nets have too many parameters to adjust (e.g., modern deep neural nets can easily have thousands or even millions of parameters that need adjusting).  

So we need a way to do this parameter adjustment automatically, regardless of the dimension of the dataset and number of parameters to adjust.  The most common way of doing this is a procedure known as **cross-validation**.  In this notebook you will learn how to use scikit-learn's built in functionality for this procedure for each of the core algorithm types. # Cross-validation for classification

In the next cell we create a familiar two class classification dataset with circular boundary. Afterwards we walk through the three prototypical algorithms we looked at earlier in the classification notebook and tune their parameters automatically using cross-validation

```python
# generate the data - here sep_x and sep_y are the coordinates of the unit circle true separator between the two classes
data, labels, true_sep = utils.make_circle_classification_dataset(num_pts = 200)
```

From the classification notebook we know that a linear classifier just won't cut it for a dataset like this.  We need something nonlinear.  So lets - one by one - tune the parameter of the nonlinear classification schemes we saw previously: a Support Vector Machine with polynomial kernel, a neural network, and a tree-based algorithim.

First we need to create a simple cross-validation function.  From lecture you saw how the basic cross-validation procedure entails


- splitting the dataset into training and testing sets (usually around 20% - 10% of the data is kept aside for the testing set)


- training a given model over a range of parameter values


- evaluating each model on the training and testing sets


- the best choice of parameter is the one providing minimum **testing error**


In the next Python cell we provide just such a function.  It takes in an arbitrary classifier, with custom parameter name and range for testing.  Study this function before using it in the cells that follow.

```python
# classification cross-validation function
from sklearn.model_selection import train_test_split
def simple_cross_validation(data,labels,clf,param_name,param_range):
    # here clf = a user-defined classification algorithm, param_name = a parameter of clf, and param_range = user-defined range of parameters of type param_name
    # split data into training and testing sets
    data_train, data_test, labels_train, labels_test = train_test_split(data, labels, test_size=0.2, random_state=42)
    
    # visualize training and testing sets
    utils.plot_train_test(data_train, data_test, labels_train, labels_test)
    
    # compute training and testing errors over range of parameters = param_range
    train_errors = list()
    test_errors = list()
    for param in param_range:
        # set chosen parameter to desired value
        exec 'clf.' + param_name + ' = ' +  str(param) in locals(),globals()  
        clf.fit(data_train, labels_train)
        train_pred = clf.predict(data_train)
        test_pred = clf.predict(data_test)

        train_accuracy = 1 - len([i for i, j in zip(labels_train, train_pred) if i == j])/float(len(labels_train))
        test_accuracy = 1 - len([i for i, j in zip(labels_test, test_pred) if i == j])/float(len(labels_test))
        train_errors.append(train_accuracy)
        test_errors.append(test_accuracy)
        
    # return training and testing errors
    return train_errors, test_errors
```

One more function - optional but helpful for visualization purposes: a small plotting function that will plot both training and testing errors together.

```python
# plot training and testing errors
def plot_cv_scores(train_errors,test_errors,param_range):
    # plot training and testing errors
    fig = plt.figure(figsize = (5,5))
    plt.plot(param_range,train_errors,marker = 'o',color =[1,0.8,0.5])
    plt.plot(param_range,test_errors,marker = 'o',color = [0,0.7,1])

    # clean up plot
    plt.xlim([min(param_range) - 0.3, max(param_range) + 0.3])
    plt.ylim([min(min(train_errors),min(test_errors)) - 0.05,max(max(train_errors),max(test_errors)) + 0.05]);
    plt.xlabel('parameter values')
    plt.ylabel('error')
    plt.title('cross validation errors',fontsize = 14)
    plt.legend(['training error','testing error'],loc='center left', bbox_to_anchor=(1, 0.5))
```

## Cross-validation with the Support Vector Machine classifier

Alright 'enough talk - its time for action'.  

First up - Support Vector Machines with a rbf kernel.  Here the parameter we need to tune is the 'gamma' - which takes on a non-negative integer value.  The param_range has been preset - but you can change it to whatever you prefer.  The same goes for the kernel type (Google 'scikit-learn SVC' to see other types of kernels you can use).

```python
# import the support vector machine classifier from scikit
from sklearn.svm import SVC

# create an instance with a polynomial kernel
clf = SVC(kernel = 'rbf')
param_name = 'gamma'
param_range = np.linspace(0,10,50)

# run our custom cross-validation function for this choice of classifier and param_name
train_errors, test_errors  = simple_cross_validation(data,labels,clf,param_name,param_range)

# plot training and testing errors
plot_cv_scores(train_errors,test_errors,param_range)
```

Now that we have computed the training and testing errors, lets choose the best one and then fit the corresponding classification model to our dataset.

```python
# pick the best parameter - i.e., the one that minimizes test error
best_param = param_range[np.argmin(test_errors)]
print 'our best cross-validated choice of parameter was: ' + param_name + ' = ' + str(best_param)

# create an instance of our classifier with the cross-validated best choice of parameter
clf = SVC(kernel = 'rbf',gamma = best_param)

# fit classifier
clf.fit(data,labels)

# plot results
utils.plot_data(data,labels,true_sep)
utils.plot_approx(clf)

# print out the number of misclassified points
predicted_labels = clf.predict(data)
acc = len(labels) - accuracy_score(labels.ravel(), predicted_labels.ravel(), normalize=False)
print 'Our classifier mislabeled ' + str(acc) + ' of ' + str(len(labels)) + ' points'
```

## Cross-validation with the neural network classifier

Up next - a neural network.  Here the parameter we need to tune is called 'hidden_layer_sizes' - which takes on a non-negative integer value.  The param_range has been preset - but you can change it to whatever you prefer.  The same goes for the parameter type and activation function type (Google 'scikit-learn neural network classifier' to see other types of parameters you can use).  Just don't change the 'solver' - because other solvers tend not to work well on small datasets like ours.

Typically in practice a neural network has a number of parameters that need tuning, and functions / architecture that need to be chosen.  But for our simple experiment a single paramter will do just fine.

```python
# import the neural network classifier from scikit
from sklearn.neural_network import MLPClassifier

# create an instance 
clf = MLPClassifier(solver = 'lbfgs')
param_name = 'hidden_layer_sizes'
param_range = np.arange(1,20)

# run our custom cross-validation function for this choice of classifier and param_name
train_errors, test_errors  = simple_cross_validation(data,labels,clf,param_name,param_range)

# plot training and testing errors
plot_cv_scores(train_errors,test_errors,param_range)
```

Now lets fit a neural network model to the dataset with minimial testing error.

```python
# pick the best parameter - i.e., the one that minimizes test error
best_param = param_range[np.argmin(test_errors)]
print 'our best cross-validated choice of parameter was: ' + param_name + ' = ' + str(best_param)

# create an instance of our classifier with the cross-validated best choice of parameter
clf = MLPClassifier(hidden_layer_sizes = best_param, solver = 'lbfgs')

# fit classifier
clf.fit(data,labels)

# plot results
utils.plot_data(data,labels,true_sep)
utils.plot_approx(clf)

# print out the number of misclassified points
predicted_labels = clf.predict(data)
acc = len(labels) - accuracy_score(labels.ravel(), predicted_labels.ravel(), normalize=False)
print 'Our classifier mislabeled ' + str(acc) + ' of ' + str(len(labels)) + ' points'
```

Last - but certainly not least - lets perform this exercise using a tree-based algorithm (here we use the excellent gradient boosting scheme).  Here the parameter we will be tuning is called 'max_depth'.  Google 'scikit-learn + gradient boosting' to see other parameters can be tuned as well.  Make sure to keep the parameter 'n_estimators' fixed at 3 for these experiments.

```python
# import the gradient boosting classifier from scikit
from sklearn.ensemble import GradientBoostingClassifier

# create an instance 
clf = GradientBoostingClassifier(n_estimators = 3)
param_name = 'max_depth'
param_range = np.arange(1,20)

# run our custom cross-validation function for this choice of classifier and param_name
train_errors, test_errors  = simple_cross_validation(data,labels,clf,param_name,param_range)

# plot training and testing errors
plot_cv_scores(train_errors,test_errors,param_range)
```

Now lets fit a gradient boosting model to the dataset with minimial testing error.

```python
# pick the best parameter - i.e., the one that minimizes test error
best_param = param_range[np.argmin(test_errors)]
print 'our best cross-validated choice of parameter was: ' + param_name + ' = ' + str(best_param)

# create an instance of our classifier with the cross-validated best choice of parameter
clf = GradientBoostingClassifier(n_estimators = 3,max_depth = best_param)

# fit classifier
clf.fit(data,labels)

# plot results
utils.plot_data(data,labels,true_sep)
utils.plot_approx(clf)

# print out the number of misclassified points
predicted_labels = clf.predict(data)
acc = len(labels) - accuracy_score(labels.ravel(), predicted_labels.ravel(), normalize=False)
print 'Our classifier mislabeled ' + str(acc) + ' of ' + str(len(labels)) + ' points'
```
